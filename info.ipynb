```python
import xarray as xr

# 1. 打开数据集
ds = xr.open_dataset('F:\\TS_PY\\KAZ_T2M_2000_2024_daily\\clip_nc\\era5_2000_clip.nc')

# 2. 查看数据集基本信息，其中会列出经纬度的范围和长度
print(ds)

# 3. 直接获取经纬度变量，并查看其最小值与最大值
longitude = ds.lon  # 坐标变量名通常是 'lon', 'longitude', 'x'
latitude = ds.lat   # 坐标变量名通常是 'lat', 'latitude', 'y'

lon_range = (longitude.min().values, longitude.max().values)
lat_range = (latitude.min().values, latitude.max().values)

print(f"经度范围: {lon_range}")
print(f"纬度范围: {lat_range}")
```
